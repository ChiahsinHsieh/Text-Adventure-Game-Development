```python
import time
import sys
#from sys import exit
import random


#Once player diceded to play, def start() to continue
def start():
    print()
    print('♦' * 73)
    
    Story_summary = f"""
    STORY SUMMARY:    

    Chia was always bullied at school by her classmate.      
    One day, she accidentally fell down the stair while avoiding chase 
    by her bully. The fell was deadly.    
    
    Before she died, 
    Her mind was fell with resentment,so she cursed the bully should also 
    experience the pain she has suffered...

    In this game, you are the bully who bullied Chia. 
    Because of the curse, your soul will enter Chia's body 
    and go to Hult to experience the bullied life that she suffered before.
    
    """
    
    for character in Story_summary:
        sys.stdout.write(character)
        sys.stdout.flush()
        time.sleep(0.02)   
    print(f"""------------------------- Game start! -------------------------""")
    input(prompt = "☞ Press Enter to Continue! ")
   
    player_name() #player input their name
    
    
#Player will input their name and start the story


print('♦' * 73)
def player_name():
    player_name = input(f"""
    Hi! Player, 
    I know you are a good person, but due to the plot needs, 
    you are the bully now!
    
    ➼ What is your name:
    >""")
    print('♦' * 73)    
    
    start = f"""
    Your name is {player_name},   
    When you open your eyes this morning,    
    you found that you have entered the body of Chia, 
    the person who you bullied...

    [Suddenly, someone speaking in your mind...]     

    :Well,well,well.         
    I said I would make you as painful as I was, this is my curse!    
    If you want your own body back,
    then start to do something good deeds!
    """
    
    for character in start:
        sys.stdout.write(character)
        sys.stdout.flush()        
        time.sleep(0.02)
    input(prompt = "☞ Press Enter to Continue! ")
    print()
    print('♦' * 73)
    Go()

#Go, ask player if they are ready to go to school.
def Go():
    while True:
        ready = input(f"""
    ➼ Are you ready to go to Hult? (Y/N)
    >""")
        
        if ready == 'Yes' or ready == 'YES' or ready =='yes' or ready == 'y' or ready == 'Y' or ready == 'sure' or ready == 'SURE' or ready == 'Sure':
            print(f"""
    ➼ You are on the way to Hult now.
        """)
            time.sleep(2)
            level1() #if player say yes, into level2
            #break
            
        elif ready == 'No' or ready == 'no' or ready == 'nope' or ready == 'n' or ready == 'N' or ready == 'Nah' or ready == 'nah' or ready == 'Nope' or ready == 'nope' :
            print(f"""
    ⚠⚠⚠ Please enter Yes or the game cannot continue!!!! ⚠⚠⚠""")

        else:
            print(f"""
        ✖✖✖ Invalid entry. Please try again! ✖✖✖""")
   
        
#level, ask player they want to use which way to enter school      
def level1():
    #while True:
        help = input(f"""
    ➼ you have reached the school gate.
    Your task is to be nice the whole day. You're now at level 1.
    There's disabled student asking for help to climb the stairs, what will you do?:

        A) Help him by yourself
        B) Ask someone to help him with you
        C) Ignore him
        >""") 
        
        if help == 'a' or help == 'A' or help == '1':
            print(f"""
    ➼ No! You can't help him along! That's too dangerous! 
    Let's try this level again!
        """)
            time.sleep(2)
            level1() #back to level1
           # break
    
        elif help == "b" or help =="B" or help == '2':
            print(f"""
    ➼ Good job! Because you helped the disabled student with other guys, 
    the student wasn't injury! Let's go to level 2 and do more good things!
    """)
            time.sleep(2)
            level2() #go to level2
            
        elif help == "c" or help =="C" or help == '3':
            print(f"""
    ➼ You are the really bad person!!! I'm forcing you back to level 1 
    do something good!
    """)
            time.sleep(2)
            level1() #back to level1
            
        else:
            print(f"""
        ✖✖✖ Invalid entry. Please try again! ✖✖✖""")
            
#level2, Ask player if they want to help the teacher
def level2():
    #while True:
        teacher = input(f"""
    ➼ Now you're on level 2! The end of the class day, the teacher asks 
    if you would stay and help him organized everyone's assignment.
    
    Would you choose help the teacher or not?
     
        A) Sure, I hope to avoid the peak crowd after school and 
           avoid encountering the bully.
        B) No, sorry I still have a lot of assignments I need to do.
        C) well...I can't. Because my mom asked me to go home early and 
           help her clean the house.
        >""")
        
        if teacher == 'A' or teacher == 'a'or teacher == '1':
            print(f"""
    ➼ You messed up the job! The teacher was so angry and asked you to leave!
    You should try it again or you can't into level 3!
        """)
            time.sleep(2)
            level2() # back to level 2
           # break

        elif teacher == 'B' or teacher == 'b'or teacher == '2':
            print(f"""
    ➼ What...? Do the assignments? I told you that you have to do something 
    good, right? Answer it again! Quick!""")
            time.sleep(2)
            level2() # back to level 2
            
        elif teacher == 'C' or teacher == 'c'or teacher == '3':
            print(f"""
    ➼ After you go home, your mom was happy that are home early and help,
    because you have never come back that early.(She was too busy to be 
    bullied by you before) 
    Anyways, well done! Let's go to level 3!""")
            time.sleep(2)
            level3()
            
        else:
            print(f"""
        ✖✖✖ Invalid entry. Please try again! ✖✖✖""")
            
#Level 3
def level3():
    #while True:
        money = input(f"""
    ➼ Next day, when you arrived at school. You see there is $20 bill on the 
    elevator floor. Will you pick it up?
    
        A)No! That's not my money!
        B)Yes! Maybe I should pick it to school security.
        C)Sure! I can use the $20 to buy dinner.
        D)Why not? It's free money!
        >""")
        
        if money == 'A' or money == 'a' or money == '1':
            print(f"""
    ➼ Hey! You should pick it up! Maybe someone is looking for their money
    and you can help them. Try to choose another answer!
    """)
            time.sleep(2)
            level3() #play dice
           # break
            
        elif money == 'B' or money == 'b' or money == '2':
            print(f"""
    ➼ What? Why would you take it to the campus security? Maybe someone is 
    on the way to back the elevator to looking for their money. 
    Put the money back!
    """)
            time.sleep(2)
            level3() #play dice
           # break
            
        if money == 'C' or money == 'c' or money == '3':
            print(f"""
    ➼ For dinner? No way! This is not your money! Choose another answer, 
    you the bad person!
    """)
            time.sleep(2)
            level3() #play dice
           # break
            
        if money == 'D' or money == 'd' or money == '4':
            print(f"""
    ➼ Alright, I will forgive you this time. 
    Because we will use the money later!
    """)
            time.sleep(2)
            level4() #play dice
            #break
            
        else:
            print(f"""
        ✖✖✖ Invalid entry. Please try again! ✖✖✖""")
            

#Level 4         
def level4():
    print(f"""
        Ok, I think it's time.""") 
    time.sleep(2)
    print(f"""
        Wait, give me the money first.""") 
    time.sleep(2)
    print(f"""
        Yes, the $20 bucks.""") 
    time.sleep(2)
    print(f"""
        Ok. You just bought the ticket to into the last challenge! 
        So I have been lying to you. 
        I don't care if you do something good ok?
        Before I leave you, 
        I will now place your fate in you own hand.
        
        The final challenge is...
    """)
    time.sleep(2)
    print(f"""
        Dice!!""")
    time.sleep(1)
    print(f"""
        Once you roll a number bigger than mine, you win!
        But remember! You only get one chance!""")
    time.sleep(1)
    print(f"""
        You ready??""")
    input(prompt = "☞ Press Enter to roll the dice! ")
    
   
    player_roll = random.randint(1,6)
    print("You rolled\n" + str(player_roll))
     
    chia_roll = random.randint(1,6)
    print("Chia rolls...." )
    time.sleep(2)
    print("Chia has rolled a\n" + str(chia_roll) )
    
    if player_roll > chia_roll:
        time.sleep(2)
        print(f"""
        Oh...you are really lucky!
        Hmmm.......
        Maybe I should make this dice game harder!!

        Anyways, you won!
        """)
            
        input(prompt = "☞ Press Enter to Continue! ")
        restart_1()

    else:
        print(f"""
        Haha! You failed! Serves you well!""")
        time.sleep(2)
        print(f"""
        You will be stuck in body forever!""")
        time.sleep(1.5)
        print(f"""
        Bye!""")
        fail()
    
    
def fail():
    print(f"""
        I'm sorry. The game is over. You have failed the game.""")
    restart()
    
    
#restart, player "failed", ask player if they want to replay the game
def restart():
    #while True:
        restart = input("""
        Do you want to play Chia's adventure game again and try to find 
        the winning way? 
        >(Y/N)) """)
        if restart == 'Y'or restart == 'y'or restart == 'Yes'or restart == 'YES'or restart == 'yes' or restart == 'Sure' or restart == 'SURE' or restart == 'sure':
            print(f"""
        Great!!! I hope that you can win this time!""")
            time.sleep(2)
            start() # Start the game again.
            #break
        
        elif restart == 'N'or restart == 'n'or restart == 'No'or restart == 'NO'or restart == 'no' or restart == 'Nah' or restart == 'nah' or restart == 'NAH':
            game_over()
        
        else:
            print(f"""
        ✖✖✖ Invalid entry. Please try again! ✖✖✖""")
            
#restart, player "won", ask player if they want to replay the game
def restart_1():
    #while True:
        restart_1 = input("""
        Good job! This game is over. I'm glad that you won the game!  
        Do you want to play again? 
        >(Y/N)) """)
        
        if restart_1 == 'Y'or restart_1 == 'y'or restart_1 == 'Yes'or restart_1 == 'YES'or restart_1 == 'yes' or restart_1 == 'Sure' or restart_1 == 'SURE' or restart_1 == 'sure':
            print(f"""
        Great!!! Let's go!""")
            input(prompt = "☞ Press Enter to Continue! ")
            start() #Start the game again.
            
        
        elif restart_1 == 'N'or restart_1 == 'n'or restart_1 == 'No'or restart_1 == 'NO'or restart_1 == 'no' or restart_1 == 'Nah' or restart_1 == 'nah' or restart_1 == 'NAH':
            game_over() ##End this game
            
        else:
            print(f"""
        ✖✖✖ Invalid entry. Please try again! ✖✖✖""")

def game_over():
    print(f"""Thank you for your participation! Bye!""")
    input(prompt = "☞ Press Enter to restart the game. ")
    input(prompt = "☞ Don't press it if you don't wanna start! ")
    input(prompt = "☞ Ha! You really wanna start! Okay, let's go! ")
           

#Player enter space to start the game
while True: 
    enter_game = input("""Welcome, Player! Do you want to play Chia's adventure game? (Y/N)) 
>""")
    if enter_game == 'Y'or enter_game == 'y'or enter_game == 'Yes'or enter_game == 'YES'or enter_game == 'yes':
        print("Great! Let's go!!")
        input(prompt = "☞ Press Enter to Continue! ")
        start()

    elif enter_game == 'N'or enter_game == 'n' or enter_game == 'no' or enter_game == 'NO':
        print(f"""
        ☹ Alright, you really broke my heart... 
        But I will keep asking you until you say 'yes'! 
        """)

    else:
        print(f"""
    ✖✖✖ Invalid entry. Please try again! ✖✖✖""")
        

            
```
